```python
urls = [
    'https://microsoft.com/en-in/',
    'https://www.apple.com/'
]
```

```python
from langchain_community.document_loaders import UnstructuredURLLoader
loaders = UnstructuredURLLoader(urls=urls)
data = loaders.load()
```

```python
data
```

```python
# Text Splitter
from langchain.text_splitter import CharacterTextSplitter

text_splitter = CharacterTextSplitter(separator='\n',
                                      chunk_size=1000,
                                      chunk_overlap=200)


docs = text_splitter.split_documents(data)

for doc in docs:
    if 'source' not in doc.metadata:
        doc.metadata['source'] = 'default_source'

for i, doc in enumerate(docs):
    if 'source' not in doc.metadata:
        print(f"Document {i} is missing 'source' metadata.")
    else:
        print(f"Document {i} has 'source' metadata: {doc.metadata['source']}")
```

```python
docs
import json

output_file_path = './data/custom_data/docs.json'
os.makedirs(os.path.dirname(output_file_path), exist_ok=True)
docs_serializable = [{
    'page_content': doc.page_content,
    'metadata': doc.metadata
} for doc in docs]
with open(output_file_path, 'w') as f:
    json.dump(docs_serializable, f, indent=4)

print(f"Documents saved to {output_file_path}")
```
